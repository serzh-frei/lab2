## Задание 1. Создание и изменение массивов

Создайте массив NumPy размером 3×3, содержащий числа от 1 до 9. Умножьте все элементы массива на 2, а затем замените все элементы больше 10 на 0. Выведите итоговый массив.

## Задание 3. Объединение и разбиение массивов

Создайте два массива NumPy размером 1×5, заполненные случайными числами от 0 до 50.
+ Объедините эти массивы в один двумерный массив (по строкам).
+ Разделите полученный массив на два массива, каждый из которых содержит 5 элементов.

Выведите все промежуточные и итоговые результаты.

## Задание 4. Генерация и работа с линейными последовательностями

Создайте массив из 50 чисел, равномерно распределенных от -10 до 10. Вычислите сумму всех элементов, сумму положительных элементов и сумму отрицательных элементов. Выведите результаты.

```python
mas = np.random.uniform(-10, 10, 50)
print(mas, '\n')

summ = sum(mas)
sum_poz = sum(mas[mas>0])
sum_neg = sum(mas[mas<0]) 

print('Сумма = ', summ)
print('Сумма положительных = ', sum_poz)
print('Сумма отрицательных = ', sum_neg)
```

## Задание 8. Определитель и обратная матрица

Создайте случайную квадратную матрицу 3×3. Найдите и выведите:
+ Определитель этой матрицы
+ Обратную матрицу (если существует, иначе выведите сообщение, что матрица вырождена)

Используйте функции `np.linalg.det` и `np.linalg.inv`.

```python
matrix = np.random.randint(1, 20, size=(3, 3))
det = np.linalg.det(matrix)

if det == 0:
    print('Матрица вырождена')
else:
    inv = np.linalg.inv(matrix)
    print('Определитель = ', det)
    print('Oбратная матрица:\n', inv)
```

## Задание 9. Транспонирование и след матрицы

Создайте матрицу NumPy размером 4×4, содержащую случайные целые числа от
1 до 50. Выведите:
+ Исходную матрицу

## Задание 10. Системы линейных уравнений 

Решите систему линейных уравнений вида:

\begin{cases} 
2x + 3y - z = 5 \\ 
4x - y + 2z = 6 \\ 
-3x + 5y + 4z = -2 
\end{cases}

Используйте матричное представление  Ax = B , где  A  – матрица коэффициентов, x  – вектор неизвестных,  B  – вектор правой части. Решите систему с помощью `np.linalg.solve` и выведите результат.

19. **Аренда конференц-зала.** Организаторы мероприятия арендуют зал и делят стоимость между тремя группами участников: обычными, VIP и студентами. Билеты для студентов стоят в два раза дешевле обычных, а VIP-билеты — в 1.5 раза дороже обычных. Всего собрано 500 000 рублей, причем студентов было в два раза больше, чем обычных участников. Сколько билетов каждого типа продано?

Обозначим за **x** стоимость обычных билетов, тогда стоимость VIP-билетов будет составлять **1.5x**, а билетов  для студентов — **0.5x**.

Количество обычных участников обозначим за **n**, a VIP-участников — за **m**. Тогда количество студентов будет равно **2n**.

Составим уравнение:

$$x*n + 1.5*x*m + 0.5*x*2*n = 500 000$$

Получаем:

$$x*n + 1.5*x*m + x*n = 500 000$$
$$2xn + 1.5xm = 500 000$$

Данное уравнение, составленное, **согласно условиям задачи**, имеет 3 неизвестных и может иметь множество решений. 


Предположим, что VIP-участников в 2 раза меньше, чем обычных участников. Тогда **m = 0.5n**.
Получаем:

$$2xn + 1.5x*0.5n = 500 000$$
$$2xn + 0.75xn = 500 000$$
$$2.75xn = 500 000$$

Данное уравнение имеет 2 неизвестных и всё ещё имеет множество решений.

Предположим, что цена обычного билета известна и равна **x = 1000**.
Тогда:

$$2.75*1000*n = 500 000$$
$$n \approx 181$$
$$m = 0.5n = 0.5*181 \approx 90$$
$$ 2n = 2*181 = 362$$
Таким образом, **число обычных участников 181, студентов — 362, VIP-участников — 90**.

**Вывод**: в данном варианте невозможно привести задачу к построению системы линейных уравнений из-за отсутствия дополнительных условий.

30. **Рацион питания спортсмена.** Спортсмену необходимо потреблять 150 г белков, 70 г жиров и 300 г углеводов в день. Он может получать их из трех продуктов: яиц, куриного филе и овсянки. В 100 г яиц содержится 12 г белков, 10 г жиров и 1 г углеводов, в 100 г куриного филе — 25 г белков, 5 г жиров и 0 г углеводов, в 100 г овсянки — 10 г белков, 5 г жиров и 60 г углеводов. Сколько граммов каждого продукта ему нужно съесть в день?

Обозначим вес продуктов в граммах: вес яиц - **x**, вес куриного филе - **y**, вес овсянки - **z**.
Составим систему уравнений:

\begin{cases} 
12x + 25y + 10z = 15000 \\
10x + 5y + 5z = 7000 \\
x + 0y + 60z = 30000 \\
\end{cases}
Умножаем правые части уравнений на 100, потому что исходные данные даны на 100 г продукта

```python
# решение матричным методом
A = np.array([[12,25,10], 
            [10, 5, 5], 
            [1, 0, 60]])
A_inv = np.linalg.inv(A) #обратная матрица
B = [15000, 7000, 30000]
X = np.dot(A_inv, B)
print('Решение матричным методом: ')
print(X[0], 'г яиц')
print(X[1], 'г куриного филе')
print(X[2], 'г овсянки')

# Решение методом крамера
det_A = np.linalg.det(A)
A1 = A.copy()
A2 = A.copy()
A3 = A.copy()
# Заменяем столбцы на вектор B

A1[:, 0] = B   
A2[:, 1] = B  
A3[:, 2] = B  
#Вычисляем определители новых матриц

det_A1 = np.linalg.det(A1)
det_A2 = np.linalg.det(A2)
det_A3 = np.linalg.det(A3)
# находим иксы

x1 = det_A1 / det_A
x2 = det_A2 / det_A
x3 = det_A3 / det_A

print('Решение методом Крамера: ')
print(x1, 'г яиц')
print(x2, 'г куриного филе')
print(x3, 'г овсянки')

#проверка третьим способом
print('Проверка методом linalg.solve(): ')
print(np.linalg.solve(A, B))
```
